# Spotify & YouTube EDA Project
This project performs Exploratory Data Analysis (EDA) on a Spotify and YouTube dataset using Python, Pandas, Matplotlib, and Seaborn.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option('display.max_columns', None)
```

```python

# Load dataset
df = pd.read_csv('spotify_youtube.csv')  # Replace with your dataset path
df.head()
```

## --- Pandas Operations ---

```python
# Remove 'Channel' column if it exists
df = df.drop(columns=['Channel'])
```

```python
df
```

```python
# Display first 5 rows
display(df.head())
```

```python
# Basic statistics
display(df.describe())
```

```python
# Shape of dataset
print("Shape of DataFrame:", df.shape)
```

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# Value counts for a categorical column (example: 'Artist')
print("Top 5 Artists by count:")
df['Artist'].value_counts().head()
```

```python
# Find the artist with the highest average 'Views'
artist_avg_views = df.groupby('Artist')['Views'].mean().sort_values(ascending=False).head(1)
print("Q1 Artist with highest average Views:\n", artist_avg_views)
```

```python
# Find the track with the largest difference between 'Likes' and 'Comments' (if both exist)
df['Likes_Comments_Diff'] = (df['Likes'] - df['Comments']).abs()
max_diff_track = df.loc[df['Likes_Comments_Diff'].idxmax(), ['Track', 'Likes_Comments_Diff']]
print("Q2 Track with largest Likes-Comments difference:\n", max_diff_track)
```

```python
# Get the top 3 tracks per artist based on 'Views'
top_tracks_per_artist = df.groupby('Artist').apply(lambda x: x.nlargest(3, 'Views')).reset_index(drop=True).head(3)
top_tracks_per_artist 
```

```python
# Calculate the percentage share of total 'Views' for each artist
total_views = df['Views'].sum()
artist_share = (df.groupby('Artist')['Views'].sum() / total_views * 100).sort_values(ascending=False)
artist_share 
```

```python
# Fill numeric NaN with column mean, categorical NaN with mode
print(df.isnull().sum())
for col in df.columns:
    if df[col].dtype in [np.float64, np.int64]:
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

## --- NumPy Questions Based on Dataset ---

```python
# --- NumPy Questions Based on Dataset ---


# Convert Views column to NumPy array
views_array = df['Views'].to_numpy()

# Normalize the 'Views' column to 0-1 range
views_normalized = (views_array - views_array.min()) / (views_array.max() - views_array.min())
print("Normalized Views:\n", views_normalized[:10])
```

```python
# Find top 5% highest 'Views' values using NumPy

threshold = np.percentile(views_array, 95)
top_5_percent_views = views_array[views_array >= threshold]
print("Top 5% Views:\n", top_5_percent_views)
```

```python
# Compute correlation coefficient between 'Views' and 'Likes' (if Likes exists)

likes_array = df['Likes'].to_numpy()
correlation = np.corrcoef(views_array, likes_array)[0, 1]
print("Correlation (Views vs Likes):", correlation)
```

```python
# Create a moving average (window=10) of the 'Views' column
window = 10
moving_avg_views = np.convolve(views_array, np.ones(window)/window, mode='valid')
print("Moving Average of Views (first 10):", moving_avg_views[:10])
```

```python
# Find indices of the top 10 'Views' values without sorting the full array
top10_indices = np.argpartition(-views_array, 10)[:10]
print("Indices of Top 10 Views:", top10_indices)
print("Top 10 Views Values:", views_array[top10_indices])
```

## --- Matplotlib Visualizations ---

```python

# --- Matplotlib Visualizations ---
import matplotlib.pyplot as plt

# Histogram
plt.figure(figsize=(8,5))
plt.hist(df['Views'], bins=30, color='skyblue', edgecolor='red')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart (example: Top 5 Artists by count)
artist_counts = df['Artist'].value_counts().head(5)
plt.figure(figsize=(6,6))
plt.pie(artist_counts, labels=artist_counts.index, autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Top 5 Artists Distribution')
plt.show()
```

```python
# Bar chart (example: Top 5 Tracks by Views)
top_tracks = df.groupby('Track')['Views'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,5))
plt.bar(top_tracks.index, top_tracks.values, color='orange')
plt.title('Top 5 Tracks by Views')
plt.ylabel('Total Views')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line chart (example: Views trend for first 20 songs)
plt.figure(figsize=(10,5))
plt.plot(df['Views'].head(20), marker='o', color='green')
plt.title('Views Trend (First 20 Songs)')
plt.xlabel('Song Index')
plt.ylabel('Views')
plt.grid(True)
plt.show()
```

## --- Seaborn Visualizations ---

```python

# Scatter Plot (Views vs Likes if exists)
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Views', y='Likes')
plt.title('Scatter Plot of Views vs Likes')
plt.show()
```

```python

# Line Plot (Views trend for first 20 songs)
plt.figure(figsize=(10,5))
sns.lineplot(x=range(20), y=df['Views'].head(20), marker='o', color='green')
plt.title('Views Trend (First 20 Songs)')
plt.xlabel('Song Index')
plt.ylabel('Views')
plt.grid(True)
plt.show()
```

```python
# Bar Plot (Top 5 Tracks by Views)
top_tracks = df.groupby('Track')['Views'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(8,5))
sns.barplot(x=top_tracks.index, y=top_tracks.values ,palette='pastel')
plt.title('Top 5 Tracks by Views')
plt.xticks(rotation=45)
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8,5))
sns.boxplot(x='Views', data=df)
plt.title('Box Plot of Views')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(8,5))
sns.violinplot(x='Views', data=df)
plt.title('Violin Plot of Views')
plt.show()
```

```python
# Count Plot (Top 5 Artists)
if 'Artist' in df.columns:
    top_artists = df['Artist'].value_counts().head(5)
    plt.figure(figsize=(8,5))
    sns.countplot(y=df['Artist'], order=top_artists.index, palette='pastel')
    plt.title('Top 5 Artists Count')
    plt.show()
```

```python
# KDE Plot
plt.figure(figsize=(8,5))
sns.kdeplot(df['Views'], fill=True, color='k')
plt.title('KDE Plot of Views')
plt.grid(True)
plt.show()
```

```python
# Distribution Plot
plt.figure(figsize=(8,5))
sns.histplot(df['Views'], kde=True, color='blue')
plt.title('Distribution of Views')
plt.grid(True)
plt.show()
```
